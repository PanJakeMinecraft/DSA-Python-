## **Arrays**

**What are Arrays in DataStructure?**
- Linear Data Structure
- Contiguous memory location
- Access element randomly
- Can have homogenrous elements (similar elements inside an array)

**Arrays properties**
- Stores items at contiguous locations
- Offers mainly the following advantages over other data structures
- Can access elements
- It is not useful in places where we have operations like insert in the middle, delete from middle and search in a unsorted data.
- It is a fundamental and linear data structure using which we build other data structures like Stack Queue, Deque, Graph, Hash Table, etc.


**Types of arrays**
 
 One dimensional Array (1D):
 - Linear functions
 - Related to a row
 - Elements can stored in one and another array
 - Only one subscript or index used
 Eg: [1,10,25,30]


Two dimensional Array(2D):
- Similar to matrix or table
- Elements are stored inside one and another 1D array
- Dimensions depends on the number of subscript used
Eg: [[1,2,3],[4,5,6],[7,8,9]]

Create a program to find the largest element inside an array using recursive method
- Input: arr[] = [20, 10, 20, 4, 100]
- Output: 100


```python
def FindMaximum(arr, n):
    if n == len(arr) - 1:
        return arr[0]
    
    max_element = FindMaximum(arr, n-1)
    return max(arr[n-1], max_element)

if __name__ == "__main__":
    arr = [20,10,20,4,100]
    print(FindMaximum(arr, len(arr)))
```

```python
# Rotate one by one

def RotateArr(arr, time):
    n = len(arr)
    for _ in range(time):
        last_element = arr[n-1]
        for j in range(n-1, 0, -1):
            arr[j] = arr[j-1]
        arr[0] = last_element

    return arr

if __name__ == "__main__":
    arr = [1,2,3,4,5,6]
    times = 3
    print(f"Before Rotation: {arr}")
    print(f"After Rotation: {RotateArr(arr, times)}")
```

### **Insertion Sort Algorithm**
- Simple sorting algorithm that work by iterating and inserting each element of an unsorted list into it's correct position

How does the algorithm works?
- For example, if ww have 5 elements array [12,11,13,5,6]
- The first element is > second element, swap the first and second
- Then compare 5 and and the sorted elements 11,12,13. Since 5 is lesser than the sorted side. Shift 5 to the beginning
- Compare 6 with the sorted element. 6 is lesser than, so we shift 6 to the left side and compare it with 5. 6 > 5, so shift it to the right side of 5

```python
# Insertion sort an array

def InsertionSort(arr):
    for i in range(1, len(arr)):
        j = i
        while arr[j-1] > arr[j] and j > 0: 
            arr[j-1], arr[j] = arr[j], arr[j-1] # swap elements if condition met
            j -= 1 # move to left side to check previous element
    return arr

if __name__ == "__main__":
    arr = [2,5,6,1,3,4]
    print(InsertionSort(arr))
```

```python
# Insertion sort array (other method)

def InsertionSort(arr):
    for i in range(1, len(arr)):
        arr_key = arr[i]
        j = i - 1
        while j >= 0 and arr_key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = arr_key
        
    return arr

if __name__ == "__main__":
    arr = [12, 11, 13, 5, 6]
    print(InsertionSort(arr))
```

### **Selection Sort Algorithm**
- Simple in place comparison sorting algorithm that work by repeately finding the minimum element from the unsorted part of the array
- Selecting the smallest or largest element from unsorted part and swap it to first unsorted element
- Find minimum, swap and repeat the process
- Quadratic time complexity O(n^2)

```python
# Selection Sort

def SelectionSort(arr):
    for i in range(0, len(arr)-1):
        min_value = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_value]:
                min_value = j
        
        if min_value != i: # swap
            arr[min_value], arr[i] = arr[i], arr[min_value]
    
    return arr

if __name__ == "__main__":
    arr = [2,6,5,1,3,4]
    print(SelectionSort(arr))
    
```

1. Arrival with minimum speed:
- Given that the distances = [30,11,23,4,20]
- You're a bus driver given that the constant speed is k (km/h) such that
    - For every station (except the last) you must round up the time to the next full hour before moving to the next station.
    - Complete the entire trip within 6 hours

Task: Find the minimum speed k that let your route complete within the time

Recommend: Use binary search method. However, if the speed is too low, the total time exceed h. If a speed is fast enough, it might still be higher than needed so therefore we search for smaller valid speed

```python
# Solution
import math

def minSpeed(dis, h):
    left, right = 1, max(dis) * h

    def OnTime(speed):
        time = 0
        for i in range(len(dis)):
            if i == len(dis) - 1:
                time += dis[i] / speed
            else:
                time += math.ceil(dis[i] / speed)
        return time <= h

    res = right
    while left <= right:
        mid = (left + right) // 2
        if OnTime(mid):
            res = mid
            right = mid - 1
        else:
            left = mid + 1
    return res
    
if __name__ == "__main__":
    dis = [30, 11, 23, 4, 20]
    h = 6
    print(f"Minimum speed: {minSpeed(dis, h):.2f}")
```

5. Union and Intersection of sorted arrays (use merge sort method)
- Given 2 arrays, a = [1,2,3,2,1] and b = [3,2,2,3,3,2], find the intersection and union of both the arrays in any order (use OOP method in Python programming)




```python
class SetNotation:
    def __init__(self, arrA, arrB):
        self.arrA = self.MergeSort(arrA)
        self.arrB = self.MergeSort(arrB)

    def MergeSort(self, arr):
        if len(arr) <= 1:
            return arr
        
        mid = len(arr) // 2
        left_half = self.MergeSort(arr[:mid])
        right_half = self.MergeSort(arr[mid:])
        return self.merge(left_half, right_half)
    
    def merge(self, left, right):
        merge_res = []
        i = j = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                merge_res.append(left[i])
                i += 1
            else:
                merge_res.append(right[j])
                j += 1
        
        merge_res.extend(left[i:])
        merge_res.extend(right[j:])
        return merge_res

    def Union(self):
        union_res = []
        i = j = 0

        while i < len(self.arrA) and j < len(self.arrB):
            if self.arrA[i] < self.arrB[j]:
                if not union_res or union_res[-1] != self.arrA[i]:
                    union_res.append(self.arrA[i])
                i += 1
            elif self.arrA[i] > self.arrB[j]:
                if not union_res or union_res[-1] != self.arrB[j]:
                    union_res.append(self.arrB[j])
                j += 1
            else:
                if not union_res or union_res[-1] != self.arrA[i]:
                    union_res.append(self.arrA[i])
                i += 1
                j += 1
        
        while i < len(self.arrA):
            if union_res[-1] != self.arrA[i]:
                union_res.append(self.arrA[i])
            i += 1
        
        while j < len(self.arrB):
            if union_res[-1] != self.arrB[j]:
                union_res.append(self.arrB[j])
            j += 1

        return union_res

    def Intersection(self):
        result = []
        i = j = 0

        while i < len(self.arrA) and j < len(self.arrB):
            if self.arrA[i] < self.arrB[j]:
                i += 1
            elif self.arrA[i] > self.arrB[j]:
                j += 1
            else:
                if not result or result[-1] != self.arrA[i]:
                    result.append(self.arrA[i])
                i += 1
                j += 1
                
        return result

if __name__ == "__main__":
    arrA = [1,2,3,2,1]
    arrB = [3,2,2,3,3,2]
    set = SetNotation(arrA, arrB)
    print(set.Union())
    print(set.Intersection())
```

Example output 2:

- Explanation: 1 moves to 3, 2 to 2, 3 to 0, 4 to 1.

- Output: 40 60 90 50 70

```python
# Reorder array given index

def ArrayMove(arr, index):
    res = [0] * len(arr)
    for i in range(len(arr)):
        res[index[i]] = arr[i]

    for i in range(len(arr)): # copy back to the array
        arr[i] = res[i]
    
    return arr

if __name__ == "__main__":
    arr = [10,11,12]
    index = [1,0,2]
    print(ArrayMove(arr, index))
```

7. Array Inversion count (with merge sort)

Given the integer array of the size n, find the inversion count in the array. Two array elements arr[i] andd arr[j] form an inversion if arr[i] > arr[j] and i < j.

Reminder: Inversion count from an array indicates how far or close the array is being sorted. Then the inversion count is 0. If the array is sorted in reverse order, the inversion count is maximum

Example output 1:
- Input: arr[] = {4, 3, 2, 1}
- Output: 6
- 3 + 2 + 1 = 6

Example output 2:
- Input: arr[] = {1, 2, 3, 4, 5}
- Output: 0
- There is no pair of indexes (i, j) exists in the given array such that arr[i] > arr[j] and i < j

**Use count and merge method O(n * logn) time complexity**

```python
# Apply the merge sort method into inversion count

class InvMergeSort:
    def __init__(self, arr):
        self.arr = arr
        self.n = len(arr)
    
    def count_inversion(self):
        temp = [0] * self.n
        return self._merge_sort(0, self.n - 1, temp)

    def _merge_sort(self, left, right, temp):
        inv_count = 0
        if left < right:
            mid = (left + right) // 2
            inv_count += self._merge_sort(left, mid, temp)
            inv_count += self._merge_sort(mid + 1, right, temp)
            inv_count += self._merge(left, mid, right, temp)  # â† Fixed this line!
        return inv_count
    
    def _merge(self, left, mid, right, temp):
        i = left
        j = mid + 1
        k = left
        inv_count = 0

        while i <= mid and j <= right:
            if self.arr[i] <= self.arr[j]:
                temp[k] = self.arr[i]
                i += 1
            else:
                temp[k] = self.arr[j]
                inv_count += (mid - i + 1)  # All remaining elements in left part
                j += 1
            k += 1
        
        while i <= mid:
            temp[k] = self.arr[i]
            i += 1
            k += 1
        while j <= right:
            temp[k] = self.arr[j]
            j += 1
            k += 1
        
        for i in range(left, right + 1):
            self.arr[i] = temp[i]
        
        return inv_count

if __name__ == "__main__":
    arr = [4, 3, 2, 1]
    inv = InvMergeSort(arr)
    print(f"Inversion count: {inv.count_inversion()}") 
```

8. Minimum swap to sort an array

Given an array arr[] of distinct elements. Find the minimum number og swaps require to sort the arrays

Example output 1:
- Input: arr[] = [2, 8, 5, 4]
- Output: 1
- Explanation: Swap 8 with 4 to get the sorted array.
 
Example output 2:
- Input: arr[] = [10, 19, 6, 3, 5]
- Output: 2
- Explanation: Swap 10 with 3 and 19 with 5 to get the sorted array.

Example output 3:
- Input: arr[] = [1, 3, 4, 5, 6]
- Output: 0
- Explanation: Input array is already sorted.

```python
# Minimum array swap

class SwapSort:
    def __init__(self, arr):
        self.arr = arr

    def MinSwap(self):
        n = len(self.arr)
        sorted_arr = self.arr[:]
        for i in range(n):
            min_idx = i
            for j in range(i+1, n):
                if sorted_arr[j] < sorted_arr[min_idx]:
                    min_idx = j
            sorted_arr[i], sorted_arr[min_idx] = sorted_arr[min_idx], sorted_arr[i]

        idx_map = {value: idx for idx, value in enumerate(sorted_arr)}
        visited = [False] * n
        swapping = 0

        for i in range(n):
            if visited[i] or idx_map[self.arr[i]] == i:
                continue
            
            cycle_size = 0
            j = i
            while not visited[j]:
                visited[j] = True
                j = idx_map[self.arr[j]]
                cycle_size += 1

            if cycle_size > 1:
                swapping += cycle_size - 1

        return swapping

if __name__ == "__main__":
    arr = [2,8,5,4]
    swap = SwapSort(arr)
    print(f"Minimum swap of array = {swap.MinSwap()}")
    
```
